```python
import numpy as np
from scipy import integrate, optimize
from scipy.misc import derivative

# Definicja funkcji g(t) podcałkowej
def integrand(t, x):
    return np.exp(x**2) / np.sqrt(1 - t**2)

# Definicja funkcji g(x) jako całki o zmiennej granicy
def g(x):
    result, _ = integrate.quad(integrand, 0, 2, args=(x,))
    return result

# Definicja pochodnej funkcji g(x)
def g_prime(x):
    return derivative(g, x, dx=1e-6)

# Znalezienie miejsc zerowych pochodnej funkcji g(x) w przedziale [0, 2]
extremum_x = optimize.root_scalar(g_prime, bracket=[0, 2])

# Obliczenie wartości funkcji g(x) w miejscu ekstremum
extremum_value = g(extremum_x.root)

print("Miejsce ekstremum:", extremum_x.root)
print("Wartość funkcji g(x) w miejscu ekstremum:", extremum_value)
```

```python
import numpy as np
from scipy import integrate, optimize

# Definicja funkcji g(t) podcałkowej
def integrand(t, x):
    return np.exp(x**2) / np.sqrt(1 - t**2)

# Definicja funkcji g(x) jako całki o zmiennej granicy
def g(x):
    result, _ = integrate.quad(integrand, 0, 2, args=(x,))
    return result

# Utworzenie funkcji celu do minimalizacji
objective_function = lambda x: -g(x)

# Znalezienie ekstremum funkcji g(x) w przedziale [0, 2]
result = optimize.minimize_scalar(objective_function, bounds=(0, 2), method='bounded')

# Miejsce ekstremum
extremum_x = result.x

# Obliczenie wartości funkcji g(x) w miejscu ekstremum
extremum_value = g(extremum_x)

print("Miejsce ekstremum:", extremum_x)
print("Wartość funkcji g(x) w miejscu ekstremum:", extremum_value)
```

```python
from scipy.integrate import quad
from scipy.optimize import minimize_scalar
import numpy as np

# Definicja funkcji podcałkowej f(t)
def f(t):
    return np.cos(t**2)

# Definicja funkcji -f(t)
def f2(t):
    return -np.cos(t**2)   

# Obliczenia wartości całki z -f(t)
def integral(x):
    result , _ = quad(f2, 0, x)
    return result

# Zdefiniowanie wartości x z przedziału x=[0.0, 2.0]
x = np.linspace(0, 2, 201)

# Dolna granica x
lower_limit = 0

# Tablice wartości całek g(x) oraz -g(x) potrzebne do wykresu
g = []
minus_g = []

# Obliczenia wartości całek dla x do wykresu
for u in x:
    result, error = quad(f, lower_limit, u)
    result2, error2 = quad(f2, lower_limit, u)
    g.append(result)
    minus_g.append(result2)
    
# Obliczenie minimum -g(x)
min_result = minimize_scalar(integral, bounds=(lower_limit, 2), method = 'bounded')
minimum_x = min_result.x
minimum_value = min_result.fun

# Rysowanie wykresu
plt.plot(x, g, label='g(x)')
plt.plot(x, minus_g, label='-g(x)')

# pionowa linia pomocnicza
plt.axvline(minimum_x, color='red', linestyle='--')

# Punkty ekstremów
plt.scatter(minimum_x, -minimum_value, color='cyan', label='Max g(x)')
plt.scatter(minimum_x, minimum_value, color='red', label='Min -g(x)')

# Opis wykresu
plt.title('Wykres funkcji $g(x)= \int_{0}^{2} \cos(t^2) dt$  oraz $-g(x)=\int_{0}^{2} - \cos(t^2) dt$')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()
plt.grid(True)
plt.show()

# Wyświetlenie wyniku
print(f'Minimum dla -g(x) oraz maksimum dla g(x) występuje dla x = {minimum_x}')
print(f'Maksimum funkcji g(x) przyjmuje wartość g(x) = {-minimum_value}')
```
