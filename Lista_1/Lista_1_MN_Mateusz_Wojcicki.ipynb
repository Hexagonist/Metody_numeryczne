$$x_{n+1}=3.5x_n(1-x_n),\qquad n=0,\ 1,\ 2,...$$

#### Zadanie 3:

Wykorzystaj moduły *numpy* i *scipy* do działań na niżej podanych macierzach
i wektorze.

$$A=\begin{pmatrix}4 & -2 & 1\\-2 & 4 & -2\\1 & -2 & 4\end{pmatrix}\,,\quad B=\begin{pmatrix}4 & 2 & 0\\2 & 5 & 2\\1 & -2 & 4\end{pmatrix}\,,\quad w=\begin{pmatrix}1 \\-2\\4\end{pmatrix} $$

Oblicz $AB$, $Aw$, $B(Aw)$ oraz wyznaczniki i macierze odwrotne $A$ i $B$.

#### Rozwiązanie:

```python
# Lists
A0 = [[4, -2, 1], [-2, 4, -2], [1, -2, 4]]
B0 = [[4, 2, 0], [2, 5, 2], [1, -2, 4]]
w0 = [[1], [-2], [4]]

# Converted lists to numpy's matrices
A = np.array(A0)
B = np.array(B0)
w = np.array(w0)

AB = np.dot(A, B)
Aw = np.dot(A, w)
B_Aw_ = np.dot(B, np.dot(A, w))

detA = np.linalg.det(A)
detB = np.linalg.det(B)

# Inversed
invA = np.linalg.inv(A)
invB = np.linalg.inv(B)


print(AB, Aw, B_Aw_, detA, detB, invA, invB, sep="\n\n")
```

#### Zadanie 4:

Elementy macierzy Hilberta $n\times n$ określone są następująco:

$$h_{ik}=\frac{1}{i+k+1}\,,\quad i,k=0,\ 1,...,\ n-1\,.$$

Wypisz elementy tej macierzy oraz macierzy odwrotnej dla $n$=4 i 8.  

Sprawdź jak zmienia się wyznacznik tej macierzy dla $n$ od 5 do 20.

#### Rozwiązanie:

```python
# Function to create Hilbert's matrix using py lists
def Hilbert_creator(r0, rn):
    h = []
    h_row = []
    for i in range(r0, rn):
        for k in range(r0, rn):
            h_row.append(1/(i+k+1))
        h.append(h_row)
        h_row = []
    return h

# Hilbert's matrices
h4 = np.array(Hilbert_creator(0, 4))
h8 = np.array(Hilbert_creator(0, 8))

# Inversed
inv_h4 = np.array(np.linalg.inv(h4))
inv_h8 = np.array(np.linalg.inv(h8))

# Results:
print("Matrices: \n")
print(h4, h8, inv_h4, inv_h8, sep="\n\n")
print("\n\n\nDeterminants: \n")

for x in range(5, 21):
    print(f"det(H[{x}x{x}]) = ", np.linalg.det(Hilbert_creator(0, x)), "\n")
```
